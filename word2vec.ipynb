#Dataset (Custom Text)


```python
text = """Modern technology is transforming the way people communicate learn and work across the world.
Students today rely on digital platforms to access educational resources attend online classes and collaborate with peers.
The availability of information has increased significantly allowing learners to explore topics beyond traditional textbooks.
Teachers are also adapting to these changes by incorporating multimedia tools interactive lessons and data driven methods to enhance student engagement.

In the business world companies are using advanced analytics and machine learning techniques to understand customer behavior and improve decision making processes.
Organizations collect large volumes of data from various sources such as social media websites and transactions.
This data is analyzed to identify patterns predict trends and optimize operations.
As a result businesses can provide personalized services and improve customer satisfaction.

Healthcare is another sector that has benefited greatly from technological advancements.
Doctors and researchers use artificial intelligence to detect diseases at early stages and recommend effective treatments.
Wearable devices help individuals monitor their health in real time tracking metrics like heart rate physical activity and sleep patterns.
Telemedicine has made it easier for patients to consult doctors remotely reducing the need for physical visits and improving accessibility in rural areas.

Transportation systems are evolving with the introduction of electric vehicles and autonomous driving technologies.
Governments and private companies are investing in sustainable solutions to reduce pollution and improve urban mobility.
Smart traffic management systems help reduce congestion and enhance road safety by analyzing real time data.
Public transportation is also becoming more efficient with the use of digital ticketing and route optimization.

In the field of entertainment streaming platforms have changed how people consume media.
Users can watch movies listen to music and access content anytime and anywhere.
Content creators are leveraging digital tools to produce high quality videos podcasts and interactive experiences.
Social media platforms allow individuals to share their creativity connect with audiences and build personal brands.

Education business healthcare transportation and entertainment are all being reshaped by innovation and digital transformation.
As technology continues to evolve it is important for individuals to develop new skills and adapt to changing environments.
Lifelong learning has become essential in a world where knowledge is constantly expanding.
People must stay updated with the latest trends and tools to remain competitive in their respective fields.

Collaboration and communication play a crucial role in this modern era.
Teams often work remotely using online platforms to share ideas manage projects and achieve common goals.
Cultural diversity in global teams brings new perspectives and encourages creativity.
Effective communication helps build strong relationships and ensures successful outcomes.

Security and privacy are also important concerns in the digital age.
Organizations must protect sensitive data from cyber threats and ensure that user information is handled responsibly.
Governments are implementing regulations to safeguard digital rights and promote ethical use of technology.
Individuals should also be aware of online risks and practice safe browsing habits.

Overall the integration of technology into daily life has created numerous opportunities and challenges.
By embracing innovation and maintaining a balanced approach society can achieve sustainable growth and improve the quality of life for people around the globe.
"""
```

#Preprocessing

```python
# Lowercase + remove punctuation
text = text.lower().translate(str.maketrans('', '', string.punctuation))
```

```python
# Tokenization (sentence → words)
sentences = nltk.sent_tokenize(text)
tokenized_data = [nltk.word_tokenize(sentence) for sentence in sentences]
```

#Train Word2Vec Models

```python
# CBOW Model (sg=0)
cbow_model = Word2Vec(
    sentences=tokenized_data,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=0
)
```

```python
# Skip-Gram Model (sg=1)
skipgram_model = Word2Vec(
    sentences=tokenized_data,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=1
)
```

#Vocabulary Size

```python
print("CBOW Vocabulary Size:", len(cbow_model.wv))
print("Skip-Gram Vocabulary Size:", len(skipgram_model.wv))
```

#Word Vectors (Example)

```python
words = ["technology", "data", "learning", "health", "media"]
```

```python
print("\n--- Word Vectors (CBOW) ---")
for word in words:
    print(word, ":", cbow_model.wv[word][:5])  # first 5 values
```

```python
print("\n--- Word Vectors (Skip-Gram) ---")
for word in words:
    print(word, ":", skipgram_model.wv[word][:5])
```

#Similar Words

```python
# For word "technology"
print("\n--- Similar Words for 'technology' ---")

print("\nCBOW:")
print(cbow_model.wv.most_similar("technology", topn=5))

print("\nSkip-Gram:")
print(skipgram_model.wv.most_similar("technology", topn=5))
```

```python
# For word "data"
print("\n--- Similar Words for 'data' ---")

print("\nCBOW:")
print(cbow_model.wv.most_similar("data", topn=5))

print("\nSkip-Gram:")
print(skipgram_model.wv.most_similar("data", topn=5))
```
